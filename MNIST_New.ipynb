```python
import torch
import torch.nn as nn
from torch.optim import Adam

class MyNet(nn.Module):
    def __init__(self):
        super().__init__()
        # 1 input channel (grayscale), 32 filters, 3x3 kernel
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1)
        self.maxm = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.relu = nn.ReLU()
        self.flatten = nn.Flatten()

        # After 2 MaxPools: 28 -> 14 -> 7. So flattened size is 64 * 7 * 7
        self.linear1 = nn.Linear(64 * 7 * 7, 128)
        self.linear2 = nn.Linear(128, 3)

    def forward(self, x):
        # x starts as (batch, 1, 28, 28)
        x = self.relu(self.conv1(x)) # (batch, 32, 28, 28)
        x = self.maxm(x)             # (batch, 32, 14, 14)
        x = self.relu(self.conv2(x)) # (batch, 64, 14, 14)
        x = self.maxm(x)             # (batch, 64, 7, 7)
        x = self.flatten(x)          # (batch, 3136)
        x = self.relu(self.linear1(x))
        x = self.linear2(x)          # (batch, 3)
        return x

mod = MyNet()
```

```python
# 1. Prepare Data (Add the channel dimension)
# xx: [300, 28, 28] -> [300, 1, 28, 28]
X = xx.unsqueeze(1).float() / 255.0
Y = yy.long()

# 2. Setup Training Tools
loss_fn = nn.CrossEntropyLoss()
optimizer = Adam(mod.parameters(), lr=0.001)

# 3. Training Loop
for epoch in range(101):
   # Output: torch.Size([32, 1, 28, 28]

    optimizer.zero_grad()           # Clear old gradients
    yp = mod(X)                     # Forward pass
    ls = loss_fn(yp, Y)             # Calculate loss
    ls.backward()                   # Backpropagation
    optimizer.step()                # Update weights

    if epoch % 10 == 0:
        # Calculate accuracy for display
        _, predicted = torch.max(yp, 1)
        acc = (predicted == Y).sum().item() / Y.size(0)
        print(f"Epoch {epoch} | Loss: {ls.item():.4f} | Acc: {acc*100:.1f}%")
```
